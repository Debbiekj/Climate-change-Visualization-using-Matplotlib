```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(2,1)
ax[0].plot(seattle_weather['MONTH'], seattle_weather["MLY-TAVG-NORMAL"],)
ax[0].plot(austin_weather['MONTH'], austin_weather["MLY-TAVG-NORMAL"], linestyle = "--", color = "r"         ) 
ax[1].plot(seattle_weather['MONTH'], seattle_weather['MLY-PRCP-NORMAL'])
ax[1].plot(austin_weather['MONTH'], austin_weather['MLY-PRCP-NORMAL'], linestyle = "--", color = "green" )
ax[0].set_xlabel("Time(month)")
ax[0].set_ylabel("average temperature")
ax[1].set_xlabel("Time(month)")
ax[1].set_ylabel('MLY-PRCP-NORMAL')
plt.show()                          
                      
```
